```
import pandas as pd
from googletrans import Translator
from langdetect import detect
```

```
data = pd.read_csv('Flickr8k.token.txt', sep='\t', header=None, engine='python')
data.columns = ['Image', 'Caption']
data.shape
```

```
data.head()
```

```
translator = Translator()
df = pd.DataFrame({
    'Image': [],
    'Caption': []
})
step = 1
```

```
for i in range(0, data.shape[0], step):
    translated = data['Caption'][i:i+step].apply(translator.translate, src='en', dest='te').apply(getattr, args=('text',))
    
    while(not (translated.apply(detect) == 'te').all()):
        print('Error')
        translated.to_csv('error.token.txt', header=None, index=None, sep='\t')
        translated = data['Caption'][i:i+step].apply(translator.translate, src='en', dest='te').apply(getattr, args=('text',))
        
    
    df = df.append(pd.DataFrame({
        'Image': data['Image'].values[i:i+step],
        'Caption': translated.values
    }))
    
    df.to_csv('Flickr8kTelugu.token.txt', header=None, index=None, sep='\t')
    print(i+step)
print('Done')
```
